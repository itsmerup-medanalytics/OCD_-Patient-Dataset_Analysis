**OBJECTIVE**

To analyze demographic and clinical characteristics of OCD patients and explore patterns related to symptom severity, comorbid conditions, and treatment documentation.

**Dataset Description**

The dataset used in this project contains demographic and clinical information of patients diagnosed with Obsessive-Compulsive Disorder (OCD). It consists of 1,500 records (rows) and 17 variables (columns).

The dataset includes demographic attributes such as age, gender, ethnicity, marital status, and education level, duration of symptoms, previous psychiatric diagnoses, family history of OCD, obsession and compulsion types, and Y-BOCS severity scores,comorbid conditions such as depression and anxiety and prescribed medication categories.This dataset is used to perform exploratory data analysis to understand demographic distributions, clinical characteristics, and symptom severity patterns among OCD patients.


```python
#Importing libraries and Loading Dataset
import pandas as pd
df=pd.read_csv('/content/OCD Patient Dataset_ Demographics & Clinical Data (2).csv')

df.head()
```

Displayed the first five rows of the dataset to understand its structure and variables.

```python
## Initial Data Exploration
df.info()
df.describe()
```

```python
## Checking for Missing Values
df.isnull().sum()
```

```python
## Handling Missing Values
df['Previous Diagnoses'] = df['Previous Diagnoses'].fillna('Not Available')
df['Medications'] = df['Medications'].fillna('Not Available')
```

```python
df.isnull().sum()
```

```python
#Checking for duplicate values
df.duplicated().sum()
```

```python
#Importing Libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
bins = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['<18', '18–25', '26–35', '36–45', '46–55', '56–65', '65+']

df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Age Group', data=df)
plt.title('Age Group Distribution of OCD Patients')
plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Gender', data=df)
plt.title('Gender-wise distribution of OCD patient')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

df['Gender'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Ethnicity', data=df)
plt.title('Ethnicity Distribution')
plt.xticks(rotation=30)
plt.show()

df['Ethnicity'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Marital Status', data=df)
plt.title('Marital Status Distribution')
plt.xticks(rotation=30)
plt.show()

df['Marital Status'].value_counts()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Education Level', data=df)
plt.title('Education Level Distribution')
plt.xticks(rotation=30)
plt.show()

df['Education Level'].value_counts()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Previous Diagnoses', data=df)
plt.title('Previous Psychiatric Diagnoses')
plt.xticks(rotation=30)
plt.show()

df['Previous Diagnoses'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Family History of OCD', data=df)
plt.title('Family History of OCD')
plt.show()

df['Family History of OCD'].value_counts()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(y='Obsession Type', data=df)
plt.title('Types of Obsessions')
plt.xlabel('Count')
plt.ylabel('Obsession Type')
plt.show()

df['Obsession Type'].value_counts()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(y='Compulsion Type', data=df)
plt.title('Types of Compulsions')
plt.xlabel('Count')
plt.ylabel('Compulsion Type')
plt.show()

df['Compulsion Type'].value_counts()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Depression Diagnosis', data=df)
plt.title('Depression Diagnosis Status')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Anxiety Diagnosis', data=df)
plt.title('Anxiety Diagnosis Status')
plt.show()

df['Anxiety Diagnosis'].value_counts()
```

```python
mean_scores = df[
    ['Y-BOCS Score (Obsessions)', 'Y-BOCS Score (Compulsions)']
].mean()
```

```python
plt.figure(figsize=(4,4))

plt.bar(
    ['Obsessions', 'Compulsions'],
    mean_scores.values,
    width=0.5
)

plt.ylabel('Average Y-BOCS Score')
plt.title('Comparison of Obsession and Compulsion Severity')

plt.tight_layout()
plt.show()
```

**Key Insights & Observations**

Age: Higher OCD prevalence is observed in adult and older age groups.
Gender: Male and female patients show nearly equal distribution.
Ethnicity: OCD cases are present across multiple ethnic groups.caucasian ethnic group shows more numbers.
Education Level: Patients from varied educational backgrounds are affected.
Marital Status: OCD occurs across all marital categories.
Family History: Family history of OCD is present in a significant subset.
Clinical Profile: Obsession and compulsion types vary widely.
Severity: Obsession and compulsion scores show comparable severity.
Comorbidities: Anxiety and depression commonly coexist with OCD.
Data Handling: Missing clinical information was retained as unavailable.
